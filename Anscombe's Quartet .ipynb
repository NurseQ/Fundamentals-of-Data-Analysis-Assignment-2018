# Anscombe's quartet dataset

   This dataset was created by statistician Francis Anscombe in 1973.  It is a set of four datasets that have similar or identical statistical properties [1] but vary dramatically when plotted in a graph. Anscombe created the datasets to highlight the importance of visualising graphs and to debunk flawed conventions around the use of graphs in mathematics and statistics.
   Visualization of data provides "the context necessary to make better choices and to be more careful when fitting models" [2].
        
Matejka and Fitzmaurice [3] reports that Anscombe did not show or suggest how the dataset was created or how it can be reproduced.

There are several attempts at creating an algorithm to produce datasets with similar statistics and have unidentical plots/graphs. Matejka and Fitzmaurice [3] reported that Chatterjee and Firat proposed a genetic algorithm technique that created 1,000 random datasets with identical statistics then combined and mutated with an objective function to maximize dissimilarity of the plots. Matejka and Fitzmaurice [3] also developed a method to use an existing dataset and modify it, and by repeating the process, create datasets with different graphical appearance and identical statistical properties.

#### Discussion

Anscombe's dataset demonstrates the importance of visualizing the data graphically. The summary of the statistical analyses of the datasets can be misleading on its own. The mean of the x and y are basically the same for each of the datasets, give or take values from the fourth decimal place. This is also the case for the variance for both, so the standard deviation is also identical. 

The regplot function of seaborn is used to visualize the data including the linear relationship, which is determined through regression. The best fit line drawn over the plots are also similar as the correlation coefficient of both x and y are similar. 

The selected dataframe above labeled as 'a' is the first dataset in Anscombe's quartet. The scatter plot reveals a simple linear relationship with some variance. 

The second dataframe labeled as 'b' shows a scatter plot that follows a curve. Linear regression may not be the appropriate determinant of correlation in this instance.

The third dataframe labeled as 'c' shows a scatter plot that seem to follow a linear distribution. There is one outlier though that changes the correlation coefficient lower from a 1 to 0.816 [4]. This example of how an outlier can affect data is made more obvious with visualization. 

The fourth dataframe labeled as 'd' shows another scatter plot with an outlier. The regression line is obviously thrown off by the effect of the one outlier. This demostrates clearly how the outlier, which has no discernible relationship with the other data points can influence the appearance of a high correlation.
